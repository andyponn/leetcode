## 1. Two Sum

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
You may assume that each input would have **exactly one solution**, and you may not use the same element twice.
You can return the answer in any order.

<font color="blue">Example 1</font>:
Input: nums = [2,7,11,15], target = 9<br>
Output: [0,1]<br3>
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].<br> <br>
<font color="blue">Example 2</font>:<br>
Input: nums = [3,2,4], target = 6<br>
Output: [1,2]<br><br>
<font color='blue'>Example 3:</font>
Input: nums = [3,3], target = 6<br>
Output: [0,1]
 

Constraints:<br>

- 2 <= nums.length <= 104<br>
- -109 <= nums[i] <= 109<br>
- -109 <= target <= 109<br>
- **Only one valid answer exists.**

```python
from typing import List 
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        num_in = nums[::-1]
        for i in range(len(nums)):
            needed= target-nums[i]
            #print(nums[i])
            #print(needed)
            if needed in nums and i != nums.index(needed):
                #print(i)
                #print(nums.index(needed))
                return [i,nums.index(needed)]

            elif needed in nums and i != len(nums)-num_in.index(needed)-1:
                return [i,len(nums)-num_in.index(needed)-1]
            else:
                continue

```

```python
# --- 以下是你在 VS Code 測試用的代碼 ---
if __name__ == "__main__":
    sol = Solution()  # 1. 實例化這個工具箱
    test_nums = [2, 11, 11, 15]
    test_target = 22
    
    # 2. 呼叫功能並印出結果
    result = sol.twoSum(test_nums, test_target)
    print(f"結果是: {result}")
```

```python
from typing import List 
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hashmap={}
        for i, num in enumerate(nums):
            needed = target -num
            if needed in hashmap:
                return [hashmap[needed],i]
            else:
                hashmap[num] = i
    
```

# 9. Palindrome Number

Given an integer x, return true if x is a palindrome, and false otherwise.
 

<font color="blue">Example 1</font>:

Input: x = 121<br>
Output: true<br>
Explanation: 121 reads as 121 from left to right and from right to left.<br>
<font color="blue">Example 2</font>:

Input: x = -121<br>
Output: false<br>
Explanation: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.<br>

<font color="blue">Example 3</font>:

Input: x = 10<br>
Output: false<br>
Explanation: Reads 01 from right to left. Therefore it is not a palindrome.<br>
 

Constraints:<br>
$-2^{31} <= x <= 2^{31} - 1$<br>
 

**Follow up: Could you solve it without converting the integer to a string?**

```python
from typing import List 
class Solution:
    def isPalindrome(self, x: int) -> bool:
        x=str(x)
        y=x[::-1]
        if x==y:
            return True
        else:
            return False
```
